```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Load Data
data = pd.read_csv('nvidia_stock_prices.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
```

```python
# Feature Selection (Close Price)
close_data = data[['Close']].values
```

```python
# Data Preprocessing
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(close_data)
```

```python
# Create Training and Test Sets
train_size = int(len(scaled_data) * 0.8)
train_data, test_data = scaled_data[:train_size], scaled_data[train_size:]
```

```python
def create_dataset(dataset, time_step=1):
    X, y = [], []
    for i in range(len(dataset)-time_step-1):
        X.append(dataset[i:(i+time_step), 0])
        y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(y)
```

```python
time_step = 60
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)

```

```python
# Reshape Input Data for LSTM
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
# Build LSTM Model with Input Layer
model = Sequential()
model.add(Input(shape=(X_train.shape[1], 1)))  # Define input shape
model.add(LSTM(50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(25))
model.add(Dense(1))
```

```python
# Compile the Model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the Model
model.fit(X_train, y_train, batch_size=64, epochs=100, verbose=1)
```

```python
# Predictions
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```

```python
# Inverse Transform for Evaluation
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
y_train = scaler.inverse_transform([y_train])
y_test = scaler.inverse_transform([y_test])
```

```python
# Plot the LSTM Results
plt.figure(figsize=(16,8))
plt.plot(data.index, close_data, label="Original Data")
plt.plot(data.index[time_step:len(train_predict)+time_step], train_predict, label="Train Predictions")
plt.plot(data.index[len(train_predict)+(time_step*2)+1:len(close_data)-1], test_predict, label="Test Predictions")
plt.legend()
plt.show()
```

```python
# RSI Calculation
delta = data['Close'].diff(1)
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
data['RSI'] = 100 - (100 / (1 + rs))
```

```python
# Plot RSI
plt.figure(figsize=(14,8))
plt.plot(data['RSI'], label='RSI')
plt.axhline(70, color='red', linestyle='--', label='Overbought')
plt.axhline(30, color='green', linestyle='--', label='Oversold')
plt.title('RSI Indicator')
plt.legend()
plt.show()
```

```python
# Bollinger Bands
data['MA20'] = data['Close'].rolling(window=20).mean()
data['BB_up'] = data['MA20'] + (data['Close'].rolling(window=20).std() * 2)
data['BB_down'] = data['MA20'] - (data['Close'].rolling(window=20).std() * 2)
```

```python
# Plot Bollinger Bands
plt.figure(figsize=(14,8))
plt.plot(data['Close'], label='Close Price')
plt.plot(data['BB_up'], label='Upper Band')
plt.plot(data['BB_down'], label='Lower Band')
plt.fill_between(data.index, data['BB_up'], data['BB_down'], color='gray', alpha=0.2)
plt.title('Bollinger Bands')
plt.legend()
plt.show()
```

```python
# Decompose the time series (Trend, Seasonality, Residuals)
decomposition = seasonal_decompose(data['Close'], model='additive', period=365)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
# Plot the decomposition
plt.figure(figsize=(14,10))
plt.subplot(411)
plt.plot(data['Close'], label='Original')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense, Input

# Update create_dataset to handle multivariate input
def create_dataset(dataset, time_step=1):
    X, y = [], []
    for i in range(len(dataset) - time_step - 1):
        X.append(dataset[i:(i + time_step)])  # Collect all features for each time step
        y.append(dataset[i + time_step, 0])   # Predict only the 'Close' price (target)
    return np.array(X), np.array(y)

# Assuming train_data and test_data are already preprocessed and contain multiple features
# 'features' is the number of columns (including 'Close')
# Create Training and Test Sets
time_step = 60  # You can adjust this based on your time series window
features = train_data.shape[1]  # Number of features (multivariate input)

X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)

# Reshape Input Data for LSTM with multivariate features
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], features)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], features)

# Build LSTM Model
model = Sequential()
model.add(Input(shape=(X_train.shape[1], X_train.shape[2])))  # Adjust input shape for multivariate features
model.add(LSTM(50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(25))
model.add(Dense(1))  # Output is a single prediction (target: 'Close')

# Compile the Model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the Model
model.fit(X_train, y_train, batch_size=64, epochs=100, verbose=1)

# Predictions
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Create a placeholder array for inverse transformation
train_predict_scaled = np.zeros((train_predict.shape[0], features))  # Placeholder for scaled train predictions
train_predict_scaled[:, 0] = train_predict[:, 0]  # Fill in only the 'Close' column predictions

test_predict_scaled = np.zeros((test_predict.shape[0], features))  # Placeholder for scaled test predictions
test_predict_scaled[:, 0] = test_predict[:, 0]  # Fill in only the 'Close' column predictions

# Inverse Transform for Evaluation (only the 'Close' column)
train_predict = scaler.inverse_transform(train_predict_scaled)[:, 0]  # Extract only 'Close' column
test_predict = scaler.inverse_transform(test_predict_scaled)[:, 0]

# Reshape y_train and y_test for inverse transform
y_train_reshaped = np.zeros((y_train.shape[0], features))  # Placeholder for actual values
y_train_reshaped[:, 0] = y_train  # Fill in only 'Close' values

y_test_reshaped = np.zeros((y_test.shape[0], features))
y_test_reshaped[:, 0] = y_test  # Fill in only 'Close' values

# Inverse transform actual values
y_train = scaler.inverse_transform(y_train_reshaped)[:, 0]
y_test = scaler.inverse_transform(y_test_reshaped)[:, 0]

# Plot the LSTM Results
plt.figure(figsize=(16, 8))
plt.plot(data.index, close_data, label="Original Data")
plt.plot(data.index[time_step:len(train_predict) + time_step], train_predict, label="Train Predictions")
plt.plot(data.index[len(train_predict) + (time_step * 2) + 1:len(close_data) - 1], test_predict, label="Test Predictions")
plt.legend()
plt.show()
```
